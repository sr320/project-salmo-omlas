# Explore Salmo-OMLAS data

Initialize the database (once) and load demo data for offline exploration.

```python
import sys
from pathlib import Path

ROOT = Path.cwd().parent
sys.path.insert(0, str(ROOT / "src"))

from salmo_omlas.db import connect, init_schema, table_exists
from salmo_omlas.demo_seed import seed
from salmo_omlas.config import get_db_path

conn = connect(get_db_path())
if not table_exists(conn, "genes"):
    init_schema(conn)
seed(conn)
print("DB ready at", get_db_path())
```

```python
import pandas as pd

pd.read_sql_query("SELECT COUNT(*) AS n FROM genes", conn)
```

```python
pd.read_sql_query(
    """SELECT g.symbol, g.biotype, pp.term_id, pp.name
       FROM genes g
       JOIN gene_processes gp ON gp.gene_id = g.id
       JOIN physiological_processes pp ON pp.id = gp.process_id
       LIMIT 20""",
    conn,
)
```
